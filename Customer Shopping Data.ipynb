```python
#indexed the data by shopping_mall and category

df.set_index (['shopping_mall','category'],inplace=True)
df
```

```python
#sorted the index in ascending order inm shopping mall and descending for category

df.sort_index(ascending = [True, False])
```

```python
df = df.reset_index(drop=True)
df
```

```python

df
```

```python
#total number of invoices for each category

category_count = df['category'].value_counts()
print(category_count)
```

```python
# the percentage of male and female customers

gender_percentage = df['gender'].value_counts(normalize=True) * 100
print(gender_percentage)
```

```python
# most popular payment method

popular_payment = df['payment_method'].mode()[0]
print(popular_payment)
```

```python
# average price for each category

category_price = df.groupby('category')['price'].mean()
print(category_price)
```

```python
# total revenue for each shopping mall

mall_revenue = df.groupby('shopping_mall')['price'].sum()
print(mall_revenue)
```

```python
# Group the data by shopping mall and calculate the total revenue generated by each mall

revenue_by_mall = df.groupby('shopping_mall')['price'].sum()
print(revenue_by_mall)
```

```python
# Filter the data to only include purchases made with credit cards

credit_purchases = df[df['payment_method'] == 'Credit Card']
print(credit_purchases)
```

```python
# top 3 categories by total revenue

top_categories_by_revenue = df.groupby('category')['price'].sum().nlargest(3)
print(top_categories_by_revenue)
```
